```python
import numpy as np
import openmc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hp_vap_rad = openmc.ZCylinder(r=0.71)
hp_liq_rad = openmc.ZCylinder(r=0.7875)
hp_ss_clad_radius = openmc.ZCylinder(r=0.8875)
hp_gap_radius = openmc.ZCylinder(r=0.8939)

inner_fuel_clad_radius = openmc.ZCylinder(r=0.9339)
inner_fuel_gap_radius = openmc.ZCylinder(r=0.9403)
fuel_hex = openmc.model.hexagonal_prism(edge_length=1.47825, orientation='y')
gap_hex = openmc.model.hexagonal_prism(edge_length=1.48564, orientation='y')

outer_fuel_ss_clad = openmc.model.hexagonal_prism(edge_length=1.60111, orientation='y')
```

```python
mats = openmc.Materials().from_xml()
```

```python
stainless = mats[0]
fuel = mats[1]
hp_vap_mat = mats[2]
hp_liq_mat = mats[3]
```

```python
hp_vap = openmc.Cell(name='HP vapor')
hp_vap.region = -hp_vap_rad
hp_vap.fill = hp_vap_mat
```

```python
hp_liq = openmc.Cell(name='HP liq')
hp_liq.region = -hp_liq_rad & +hp_vap_rad
hp_liq.fill = hp_liq_mat
```

```python
hp_clad = openmc.Cell(name='HP clad')
hp_clad.region = -hp_ss_clad_radius & +hp_liq_rad
hp_clad.fill = stainless
```

```python
hp_gap = openmc.Cell(name='HP gap')
hp_gap.region = -hp_gap_radius & +hp_ss_clad_radius
hp_gap.fill = None
```

```python
fuel_inner_clad = openmc.Cell(name='Fuel Inner Clad')
fuel_inner_clad.region = -inner_fuel_clad_radius & +hp_gap_radius
fuel_inner_clad.fill = stainless
```

```python
fuel_inner_gap = openmc.Cell(name='Fuel Inner Gap')
fuel_inner_gap.region = -inner_fuel_gap_radius & +inner_fuel_clad_radius
fuel_inner_gap.fill = None
```

```python
fiss = openmc.Cell(name='Fuel')
fiss.region = fuel_hex & +inner_fuel_gap_radius
fiss.fill = fuel
```

```python
fuel_outer_gap = openmc.Cell(name='Fuel Outer Gap')
fuel_outer_gap.region = gap_hex & openmc.Complement(fuel_hex)
fuel_outer_gap.fill = None
```

```python
outer_clad = openmc.Cell(name='Outer Clad')
outer_clad.region = outer_fuel_ss_clad & openmc.Complement(gap_hex)
outer_clad.fill = stainless
```

```python
outer_unit_shape = openmc.model.hexagonal_prism(edge_length=1.6085, orientation='y')
outer_unit = openmc.Cell(name='Outer Unit')
outer_unit.region = outer_unit_shape & openmc.Complement(outer_fuel_ss_clad)
outer_unit.fill = None
```

```python
fuel_pin = openmc.Universe(cells=(hp_vap, hp_liq, hp_clad, hp_gap, fuel_inner_clad, fuel_inner_gap, fiss, fuel_outer_gap, outer_clad, outer_unit))
```

```python
fuel_pin.plot(basis='xy', width=(3.0,3.0), pixels=(500, 500), color_by='material')
```

```python
all_ss_cell = openmc.Cell(fill=stainless)
outer_universe = openmc.Universe(cells=(all_ss_cell,))
```

```python
lat = openmc.HexLattice()
lat.center = (0.,0.)
lat.pitch = (2.786,)
lat.outer = outer_universe
```

```python
inner_assembly_r = openmc.ZCylinder(r=9.05)
inner_assembly_hex = openmc.model.hexagonal_prism(edge_length=11.02854, orientation='y')

inner_void = openmc.Cell(name='Inner Core')
inner_void.region = -inner_assembly_r
inner_void.fill = None

inner_hex = openmc.Cell(name='Inner Hex')
inner_hex.region = inner_assembly_hex & +inner_assembly_r
inner_hex.fill = stainless
```

```python
outer_lat_region = openmc.model.hexagonal_prism(edge_length=57.38862, orientation='y')
lattice = openmc.Cell(name='lattice')
lattice.fill = lat
lattice.region = outer_lat_region & openmc.Complement(inner_assembly_hex)
assembly = openmc.Universe(cells=(lattice, inner_hex, inner_void))
```

```python
stainless_cell = openmc.Cell(name='Stainless Pin')
stainless_cell.region = outer_fuel_ss_clad
stainless_cell.fill = stainless
stainless_pin = openmc.Universe(cells=(stainless_cell, outer_unit))
```
